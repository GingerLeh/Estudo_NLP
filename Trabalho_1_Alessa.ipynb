### Segmentação das palavras

```python
palavras = nltk.tokenize.word_tokenize(texto)
```

### Definindo rotuladores e stopwords

```python
nltk.download('mac_morpho')
```

```python
# Criando um rotulador padrão
rotulador_padrao = nltk.DefaultTagger('desconhecido')
```

```python
rotulador_padrao.tag(palavras)
```

```python
# Definindo novo rotulador mais inteligente 
train = nltk.corpus.mac_morpho.tagged_sents()
rotulador = nltk.UnigramTagger(train = train, backoff=rotulador_padrao)
```

```python
tokens_rotulados = rotulador.tag(palavras)
```

```python
tokens_rotulados
```

```python
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
stopwords.extend(['.',',','?','!', '(',')', ':','-','—'])
```

```python
tokens = [token for token in palavras if token not in stopwords]
```

```python
tokens
```

```python
tokens_rotulados = rotulador.tag(tokens)
```

```python
tokens_rotulados
```

### Stemmização e Lematização

```python
nltk.download('rslp')
```

```python
stemmer = nltk.stem.RSLPStemmer()
```

```python
for sentenca in tokens:
  print([stemmer.stem(token) for token in sentenca.split(' ')])
```

#### Podemos notar que a técnica não é muito eficaz e não produz um resultado muito bom. Podemos usar agora a biblioteca spacy para melhorar o resultado e fazer uma Lematization.

```python
nlp = spacy.load("pt_core_news_lg")
```

```python
texto_spacy = nlp(texto)
```

```python
#Fazendo a lematização do texto e salvando em um arquivo o resultado para melhor visualização
novo_texto_lematizado = []
arquivo = open('novo_texto_lematizado.txt', 'w')
for token in texto_spacy: 
    if token.text != ',' and token.text != '.' and token.text != ')' and token.text != '(':
        novo_texto_lematizado.append(token)
        arquivo.write(token.lemma_ + " ")
        print(token.lemma_)
arquivo.close()
```

```python
novo_texto_lematizado[:30]
```

### Estudo de similaridade

```python
for token in texto_spacy: 
  print(token.vector)
```

```python
def similaridade_cossenos(token1, token2):
  token1 = token1.vector
  token2 = token2.vector
  return np.dot(token1,token2)/(np.linalg.norm(token1)*np.linalg.norm(token2))
```

### 3 - Qual a similaridade entre as palavras "UNESP", "USP" e "Universidade"?

```python
token_unesp = nlp("UNESP")[0]
token_usp = nlp("USP")[0]
token_universidade = nlp("Universidade")[0]
```

```python
quantidades = []
labels = ['Usp e Unesp','Unesp e Universidade','Usp e Universidade']
```

UNESP e USP

```python
quantidades.append(similaridade_cossenos(token_unesp,token_usp)*100)
```

UNESP e Universidade

```python
quantidades.append(similaridade_cossenos(token_unesp, token_universidade)*100)
```

USP e Universidade

```python
quantidades.append(similaridade_cossenos(token_usp, token_universidade)*100)
```

```python
quantidades
```

A similaridade entre Unesp e Usp é cerca de 68%, enquanto que Unesp e Universidade é cerca de 50%, e Usp e Universidade é cerca de 46%.

```python
cores = ['pink', 'orange', 'green']
nomes = ['Usp e Unesp', 'Unesp e Universidade', 'Usp e Universidade']
plt.barh(range(len(quantidades)), quantidades, color=cores)
for i, v in enumerate(quantidades):
     plt.text(v, i, "{:.2f}".format(v))

for i, nome in enumerate(nomes):
    plt.text(quantidades[i]/2, i, nome, ha='center', va='center')
plt.xlabel('Porcentagem')
plt.show()
```
