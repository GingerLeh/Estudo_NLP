```python
def similaridade_cossenos(token1, token2):
  token1 = token1.vector
  token2 = token2.vector
  return np.dot(token1,token2)/(np.linalg.norm(token1)*np.linalg.norm(token2))
```

```python
def similaridade_cosseno2(token1, token2):
  token1 = token1.vector
  token2 = token2.vector
  return (token1 @ token2)/ (np.linalg.norm(token1)*np.linalg.norm(token2))
```

```python
similaridade1 = similaridade_cossenos(token_banana, token_uva)
similaridade1
```

```python
similaridade2 = similaridade_cosseno2(token_banana, token_uva)
similaridade2
```

```python
token_mulher = nlp('mulher')[0]
token_homem = nlp('homem')[0]
```

```python
similaridade_homem_mulher = similaridade_cossenos(token_homem,token_mulher)
similaridade_homem_mulher
```

```python
token_mulher.similarity(token_homem)
```

```python
def menos_mais(token1,token2,token3): # (token1 - token2) + token3
  resultado_da_subtação = token1.vector - token2.vector
  resultado = resultado_da_subtação + token3.vector
  vetor_resultado = nlp.vocab.vectors.most_similar(np.asarray([resultado]))
  return nlp.vocab.strings[vetor_resultado[0]]
```

```python
token_rainha = nlp('rainha')[0]
```

```python
menos_mais(token_rainha, token_mulher, token_homem)
```
